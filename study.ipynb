```python
import pandas as pd
import numpy as np
```

```python
crimes_df = pd.read_csv("Crimes.csv")
moon_df = pd.read_csv("full_moon.csv")
holidays_df = pd.read_csv("holidays.csv")
weather_df = pd.read_csv("weather.csv")
```

```python
# crimes_df.head()

#convert Date to datetime
crimes_df["Date"] = pd.to_datetime(crimes_df["Date"])
#make Date into two columns with the date and time
crimes_df["Time"] = crimes_df["Date"].dt.time
crimes_df["Date"] = crimes_df["Date"].dt.date
crimes_df.head()
```

```python
# moon_df.head()
#rename FullMoonDates to Date
moon_df.rename(columns = {'FullMoonDates':'Date'}, inplace = True)
#convert Date to datetime
moon_df["Date"] = pd.to_datetime(moon_df["Date"])
moon_df.head()
```

```python
# holidays_df.head()
holidays_df['Date'] = pd.to_datetime(holidays_df['Date'])
holidays_df.head()
```

```python
# weather_df.head()
#rename datetime to Date
weather_df.rename(columns = {'datetime':'Date'}, inplace = True)
#convert Date to datetime
weather_df["Date"] = pd.to_datetime(weather_df["Date"])
weather_df.head()
```

```python
# type(crimes_df["Date"][0])
# type(moon_df["Date"][0])
# type(holidays_df["Date"][0])
# type(weather_df["Date"][0])
crimes_df['Date'] = pd.to_datetime(crimes_df['Date'])
type(crimes_df["Date"][0])
```

```python
merged_df =pd.merge(crimes_df, moon_df, on='Date', how='outer')
merged_df =pd.merge(merged_df, holidays_df, on='Date', how='outer')
merged_df =pd.merge(merged_df, weather_df, on='Date', how='outer')
# with all columns
merged_df
```
